# Read the dataset


```python
import numpy as np
import pandas as pd;
import matplotlib.pyplot as plt

data_x=pd.read_csv('xdata.csv')
data_y=pd.read_csv('ydata.csv')

plt.style.use('seaborn')
```

# Convert data to numpy array 

```python
X=data_x.values
Y=data_y.values
```

# Since we dont need the first column so slice it

```python
X=X[:,1:]
Y=Y[:,1:]

print(X)

print(X.shape)
print(Y.shape)

Y=Y.reshape((-1,))
```

# Make Scatter Plot

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

# Let 2,3 be our query point

```python

query_x=np.array([2,3])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query_x[0],query_x[1],color='Red')
plt.show()
    
```

```python
def distance(X1,X2):
    return np.sqrt(sum((X1-X2)**2))
  
def knn(X,Y,query_point,K=5):
    vals=[]
    m=X.shape[0]
    for i in range(m):
        d=distance(query_point,X[i])
       # vals.append((dist,Y[i]))
        vals.append((d,Y[i]))
    #Sort the array according to distances
    vals=sorted(vals)
    
    #Keep only k points
    vals = vals[:K]
    vals = np.array(vals)
    
   # print(vals)
    new_vals=np.unique(vals[:,1],return_counts=True)
    print(new_vals)
    
    index=new_vals[1].argmax()
  #  print(index)
    pred=new_vals[0][index]
    return pred
```

```python
vals=knn(X,Y,query_x)
print(vals)
```

```python
query_x=np.array([0,2])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query_x[0],query_x[1],color='Red')
plt.show()
```
