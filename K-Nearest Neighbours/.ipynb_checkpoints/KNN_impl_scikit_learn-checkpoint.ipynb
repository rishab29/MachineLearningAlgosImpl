```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn')

data_x=pd.read_csv('xdata.csv')
data_y=pd.read_csv('ydata.csv')

X=data_x.values
Y=data_y.values

X=X[:,1:]
Y=Y[:,1:]

Y=Y.reshape(-1)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
query_x=np.array([[0,2]])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query_x[0][0],query_x[0][1],color='Red')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_classifier=KNeighborsClassifier(n_neighbors=5,p=2)
knn_classifier.fit(X,Y)
knn_classifier.predict(query_x)
```
