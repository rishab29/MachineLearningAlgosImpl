# Linear Regression Implementation


### First Step Is to Fetch the Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset_x=pd.read_csv('linearX.csv')
dataset_y=pd.read_csv('linearY.csv')

x=dataset_x.values
y=dataset_y.values

x=x.reshape(-1,)
y=y.reshape(-1,)
print(x.shape)
print(y.shape)
```

# Second step is to Plot The initial Data

```python
plt.scatter(x,y)
X = (x-x.mean())/x.std()
Y = y
plt.scatter(X,Y)
plt.show()
```

```python
plt.scatter(X,Y)
plt.show()
```

# Calculating the Gradient Descent

# 1.Define Hypothesis

```python
def hypothesis(x,theta):
    return theta[0]+theta[1]*x;
```

# 2.Square Errored Function

```python
def error(x,y,theta):
    m=x.shape[0]
    error=0
    for i in range(m):
        hx=hypothesis(x[i],theta)
        error=error+(hx-y[i])**2
        
    
    return error
        
```

# 3.Define Gradient

```python
def gradient(x,y,theta):
    m=x.shape[0]
    grad=np.zeros((2,))
    for i in range(m):
        hx=hypothesis(x[i],theta)
        grad[0]=grad[0]+hx-y[i]
        grad[1]=grad[1]+(hx-y[i])*x[i]
        
    return grad    
```

# 4.Calculate Gradient Descent

```python
def gradientDescent(x,y,learning_rate=0.001):
    
    theta = np.array([-2.0,0.0])
    itr=1
    maxitr=100;
    error_list=[]
    theta_list=[]
    while(itr<=maxitr):
        grad=gradient(x,y,theta)
        e=error(x,y,theta)
        error_list.append(e)
        
        theta_list.append((theta[0],theta[1]))
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
        
        itr=itr+1
        
    return theta,error_list,theta_list    
    
    
```

```python
final_theta, error_list,theta_list = gradientDescent(X,Y)
```

```python
plt.plot(error_list)
plt.show()
```

```python
print(final_theta)
```

```python
### Plot the line for testing data

xtest = np.linspace(-2,6,10)
print(xtest)
```

```python
plt.scatter(X,Y,label='Training Data')
plt.scatter(xtest,hypothesis(xtest,final_theta),color='orange',label="Prediction")
plt.legend()
plt.show()
```

# Visualising Gradient Descent
Plotting Error surface and contours

```python
from mpl_toolkits.mplot3d import Axes3D
#ax = fig.add_subplot(111,project='3d')
T0 = np.arange(-2,3,0.01)
T1 = np.arange(-2,3,0.01)
T0,T1 = np.meshgrid(T0,T1)
J = np.zeros(T0.shape)
m = T0.shape[0]
n = T0.shape[1]

for i in range(m):
    for j in range(n):
        J[i,j] = np.sum((Y - T1[i,j]*X - T0[i,j])**2)
        
fig = plt.figure()
axes = fig.gca(projection='3d')

theta_list = np.array(theta_list)
axes.scatter(theta_list[:,0],theta_list[:,1],error_list,c='k')

axes.plot_surface(T0,T1,J,cmap='rainbow',alpha=.9)
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(T0,T1,J,cmap='rainbow')
axes.set_xlim([-2,2])
axes.set_ylim([-2,2])
axes.scatter(theta_list[:,0],theta_list[:,1],error_list,c='k',marker='^')
plt.title("3D Contour")
plt.show()
```
